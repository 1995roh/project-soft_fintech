```python
f,ax = plt.subplots(1,1,figsize=(5,5))
sns.countplot(x='median_house_value',data=pro)
plt.show()
```

```python
x=pro['median_house_value']
y=pro['median_income']
plt.title("Median Income")
plt.xlabel("price of house")
plt.ylabel("income")
plt.plot(x,y)
plt.hist
plt.show()
```

```python
x=pro['median_house_value']
y=pro['total_bedrooms']
plt.title("Total Bedrooms")
plt.xlabel("price of house")
plt.ylabel("no.of bedrooms")
plt.bar(x,y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(pro['median_house_value'],pro['median_income'])
ax.set_title("Scatter plot of house value")
ax.set_xlabel("price of house")
ax.set_ylabel("income")
plt.show()
```

```python
colors={'<1H OCEAN':'r','INLAND':'b','NEAR OCEAN':'y','NEAR BAY':'w','ISLAND':'g'}
fig,ax=plt.subplots()
for i in range(len(pro['median_house_value'])):
    ax.scatter(pro['median_house_value'][i],pro['median_income'][i],color=colors[pro['ocean_proximity'][i]])

ax.set_title("Scatter plot of house value")
ax.set_xlabel("price of house")
ax.set_ylabel("income")
plt.show()
```

```python
columns=pro.columns.drop(['ocean_proximity'])
x_data=range(0,pro.shape[0])
fig,ax=plt.subplots()
for c in columns:
    ax.plot(x_data,pro[c])
ax.set_title('housing dataset')
ax.legend(labels=[])
plt.show()    
```
